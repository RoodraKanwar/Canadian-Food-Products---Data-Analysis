```python
import pandas as pd
import numpy as np
import plotly.express as px
import plotly.graph_objs as go
from plotly.offline import iplot
```

```python
df = pd.read_csv('/content/BC_Food_2017_2024.csv')
```

```python
df.head(4)
```

```python
df.columns = df.iloc[0]
df = df.drop(0)
df.dropna(inplace=True)
```

```python
df
```

```python
chicken_products = df[df['Products'].str.contains('Chicken', case=False)]
chicken_products
```

```python
# Convert from wide to long format
chicken_products_melt = chicken_products.melt(id_vars=["Products"], var_name="Month", value_name="Price")
chicken_products_melt
```

```python
# Convert "Month" to datetime format
chicken_products_melt["Month"] = pd.to_datetime(chicken_products_melt["Month"], format="%B %Y")
chicken_products_melt
```

```python
whole_chicken_df = chicken_products_melt[chicken_products_melt['Products'].str.contains("Whole Chicken", case=False)]
whole_chicken_df = whole_chicken_df.sort_values(by=["Month"], ascending=True)
whole_chicken_df.head(4)
```

```python
whole_chicken_df.to_csv("Whole_chicken.csv", index=False)
```

```python
data = go.Scatter(x = whole_chicken_df['Month'],
                   y = whole_chicken_df['Price'],
                   mode = 'lines',
                   name = 'Whole Chicken',
                   marker = dict(color = 'darkorange'))

layout = go.Layout(title={'text': "Chicken Products Prices",
                          'y':0.9,
                          'x':0.5,
                          'xanchor': 'center',
                          'yanchor': 'top'},
                   template = 'plotly_dark',
                   xaxis=dict(showgrid=False, zeroline=False, linecolor='gray'),
                   yaxis=dict(showgrid=False, zeroline=False, linecolor='gray'))

fig = go.Figure(data = data, layout = layout)
iplot(fig)
```

```python
whole_chicken = go.Scatter(x = whole_chicken_df['Products'],
                     y = whole_chicken_df['Month'],
                     mode = 'lines',
                     marker = dict(color = 'darkorange'))

layout = go.Layout(title={'text': "Chicken Products Prices",
                          'y':1,
                          'x':0.5,
                          'xanchor': 'center',
                          'yanchor': 'top'},
                   template = 'plotly_dark')

fig = go.Figure(data = [whole_chicken], layout = layout)
iplot(fig)
```

```python
import pandas as pd
import plotly.express as px

# Load the data
data = pd.read_csv('Whole_chicken.csv')

# Convert the 'Month' column to datetime
data['Month'] = pd.to_datetime(data['Month'])

# Sort the data by date to ensure it's in chronological order
data = data.sort_values(by='Month')

# Create the plot using Plotly Express
fig = px.line(data, x='Month', y='Price', title='Whole Chicken Price Over Time',
              labels={'Price': 'Price (per kilogram)', 'Month': 'Date'})

# Customize the layout (optional)
fig.update_layout(
    xaxis_title='Date',
    yaxis_title='Price (per kilogram)',
    xaxis=dict(tickangle=45),  # Rotate x-axis labels for better readability
    showlegend=False,  # Hide legend if not needed
    template='plotly_white'  # Use a clean template
)

# Show the plot
fig.show()
```
