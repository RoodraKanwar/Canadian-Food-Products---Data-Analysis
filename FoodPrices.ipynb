```python
import pandas as pd
import numpy as np
import plotly.express as px
```

```python
df = pd.read_csv('/content/BC_Food_2017_2024.csv')
```

```python
df.head(4)
```

```python
df.columns = df.iloc[0]
df = df.drop(0)
df.dropna(inplace=True)
df.head(4)
```

```python
chicken_products = df[df['Products'].str.contains('Chicken', case=False)]
chicken_products
```

```python
# Convert from wide to long format
chicken_products_melt = chicken_products.melt(id_vars=["Products"], var_name="Month", value_name="Price")
chicken_products_melt
```

```python
chicken_products_melt.dtypes
```

```python
# Convert "Month" to datetime format
chicken_products_melt["Month"] = pd.to_datetime(chicken_products_melt["Month"])
chicken_products_melt['Price'] = chicken_products_melt['Price'].astype(float)
chicken_products_melt
```

```python
whole_chicken_df = chicken_products_melt[chicken_products_melt['Products'].str.contains("Whole Chicken", case=False)]
chicken_products_melt.to_csv("Whole_chicken.csv", index = False)
```

```python
whole_chicken_df = whole_chicken_df.sort_values(by='Month')
whole_chicken_df.reset_index(drop=True, inplace=True)
whole_chicken_df
```

```python
# Plot
fig = px.line(whole_chicken_df, x='Month', y='Price', title='Whole Chicken Price Over Time',
              labels={'Price': 'Price (per kilogram)', 'Month': 'Date'})

# Layout
fig.update_layout(
    xaxis_title='Date',
    yaxis_title='Price (per kilogram)',
    showlegend=False,
    template='plotly_dark',
    xaxis=dict(showgrid=False),
    yaxis=dict(showgrid=True)
)

# Show the plot
fig.show()
```
