```python
import numpy as np
import pandas as pd
import plotly.express as px
```

```python
df = pd.read_csv("/content/18100245.csv")
```

```python
df
```

```python
df['GEO'].value_counts()
```

```python
geo = ['Ontario','Alberta','British Columbia','Quebec']
filtered_df = df[df['GEO'].isin(geo)]

filtered_df = filtered_df[['REF_DATE','GEO','Products','VALUE']]
filtered_df = filtered_df.rename(columns={'REF_DATE':'Ref_date','GEO':'Province','GEO':'Province','VALUE':'Price'})
filtered_df = filtered_df.reset_index(drop=True)
filtered_df
```

```python
filtered_df['Products'].value_counts()
```

```python
pivot_df = filtered_df.pivot(index=['Province','Products'], columns='Ref_date', values='Price').reset_index()
pivot_df
```

```python
pivot_df.dtypes
```

```python
pivot_df['First Quarter Average'] =(pivot_df['2017-01'] + pivot_df['2017-02'] + pivot_df['2017-03']) / 3
pivot_df['Last Quarter Average'] =(pivot_df['2024-12'] + pivot_df['2024-11'] + pivot_df['2024-10']) / 3
pivot_df['Rate Change'] = ((pivot_df['First Quarter Average'] - pivot_df['Last Quarter Average'])/pivot_df['First Quarter Average']) * 100
pivot_df
```

```python
pivot_df = pivot_df.sort_values(by='Rate Change')
pivot_df
```

```python
red_meat = pivot_df[pivot_df['Products'].str.contains('Beef','Pork')]
red_meat
```

```python
filtered_df['Date'] = pd.to_datetime(filtered_df['Ref_date'])

beef_df = filtered_df[filtered_df['Products'].str.contains('Beef', case=False, na=False)]
pork_df = filtered_df[filtered_df['Products'].str.contains('Pork', case=False, na=False)]

beef_df = beef_df.copy()
pork_df = pork_df.copy()

beef_df['Products'] = beef_df['Products'].str.split(',').str[0]
pork_df['Products'] = pork_df['Products'].str.split(',').str[0]

beef_df.loc[:, 'Province Product'] = beef_df['Province'] + " - " + beef_df['Products']
pork_df.loc[:, 'Province Product'] = pork_df['Province'] + " - " + pork_df['Products']
```

```python
# Plot
color_palette = px.colors.qualitative.Vivid
fig = px.line(beef_df, x='Date', y='Price', title='Beef Price Trend by Province', color='Province Product',
              labels={"Price": "Price (per kilogram)", "Date": "Month"},
              color_discrete_sequence=color_palette)

# Layout
fig.update_layout(
    xaxis_title='Date',
    yaxis_title='Price (per kilogram)',
    showlegend=True,
    template='plotly_dark',
    xaxis=dict(showgrid=False),
    yaxis=dict(showgrid=True),
)

fig.show()
```

```python
# Plot
color_palette = px.colors.qualitative.Vivid
fig = px.line(pork_df, x='Date', y='Price', title='Beef Price Trend by Province', color='Province Product',
              labels={"Price": "Price (per kilogram)", "Date": "Month"},
              color_discrete_sequence=color_palette)

# Layout
fig.update_layout(
    xaxis_title='Date',
    yaxis_title='Price (per kilogram)',
    showlegend=True,
    template='plotly_dark',
    xaxis=dict(showgrid=False),
    yaxis=dict(showgrid=True),
)

fig.show()
```
